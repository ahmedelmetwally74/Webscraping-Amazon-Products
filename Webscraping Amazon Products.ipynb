```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 
import requests 
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.amazon.in/gp/bestsellers/books/")
```

```python
titles=driver.find_elements(By.XPATH,"//div[@class ='_cDEzb_p13n-sc-css-line-clamp-1_1Fn1y']")
for title in titles:
    print(title.text)
```

```python
prices=driver.find_elements(By.XPATH,"//span[@class ='_cDEzb_p13n-sc-price_3mJ9Z']")
for price in prices:
    print(price.text)
```

```python
data = []
```

```python
for title, price in zip(titles, prices):
    data.append([title.text, price.text])
```

```python
csv_file_path = "Books.csv"
```

```python
# Write data to CSV file
import csv
with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Title', 'Price'])
    csv_writer.writerows(data)
```
